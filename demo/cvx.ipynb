```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
```

```python
bound = 2
```

```python
x = cp.Variable(1)
objective = cp.Minimize(x**2)
constraint = [x >= bound]
problem = cp.Problem(objective, constraint)
```

```python
problem.solve()
print("optimal x value is %.2f"%x.value[0])
```

```python
t = np.linspace(-10,10,100)
plt.figure(figsize=(6,4))
plt.plot(t, t**2, label="objective")
plt.vlines(bound, 0, 100, color='red', label="constraint boundary")
plt.fill_betweenx([0,100], -10, bound, color='red', alpha=0.2, label="infeasible")
plt.scatter(x.value, x.value**2, marker="*", color="green", s=200, zorder=10, label="optimal")
plt.legend(ncol=2)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```
